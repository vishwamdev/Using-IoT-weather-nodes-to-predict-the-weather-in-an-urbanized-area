```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
pre1_df = pd.read_csv(r'/content/node_2data/2 Node prediction/Reference Node.csv')
pre2_df = pd.read_csv(r'/content/node_2data/2 Node prediction/Prediction with 2 node.csv')
```

```python
min_len = min(len(pre1_df), len(pre2_df))
pre1_df = pre1_df.iloc[:min_len]
pre2_df = pre2_df.iloc[:min_len]
```

```python
n = min_len
temp_relative_error = (1/n) * sum(abs(pre1_df['Temperature'] - pre2_df['Pre_Temperature']) / abs(pre1_df['Temperature']) * 100)
temp_point_errors = abs(pre1_df['Temperature'] - pre2_df['Pre_Temperature']) / abs(pre1_df['Temperature']) * 100
```

```python
hum_relative_error = (1/n) * sum(abs(pre1_df['Humidity'] - pre2_df['Pre_Humidity']) / abs(pre1_df['Humidity']) * 100)
hum_point_errors = abs(pre1_df['Humidity'] - pre2_df['Pre_Humidity']) / abs(pre1_df['Humidity']) * 100
```

```python
# Plot Temperature (Actual as X, Predicted as Y)
plt.figure(figsize=(10, 5))
plt.scatter(pre1_df['Temperature'], pre2_df['Pre_Temperature'], c='blue', alpha=0.5)
plt.plot([pre1_df['Temperature'].min(), pre1_df['Temperature'].max()],
         [pre1_df['Temperature'].min(), pre1_df['Temperature'].max()], 'r--')
plt.xlabel('Actual Temperature')
plt.ylabel('Predicted Temperature')
plt.title('Temperature: Actual vs Predicted')
plt.grid(True)
plt.show()
```

```python
# Plot Humidity (Actual as X, Predicted as Y)
plt.figure(figsize=(10, 5))
plt.scatter(pre1_df['Humidity'], pre2_df['Pre_Humidity'], c='green', alpha=0.5)
plt.plot([pre1_df['Humidity'].min(), pre1_df['Humidity'].max()],
         [pre1_df['Humidity'].min(), pre1_df['Humidity'].max()], 'r--')
plt.xlabel('Actual Humidity')
plt.ylabel('Predicted Humidity')
plt.title('Humidity: Actual vs Predicted')
plt.grid(True)
plt.show()
```

```python
# Print relative errors
print(f"Total Relative Error for Temperature: {temp_relative_error:.2f}%")
print(f"Total Relative Error for Humidity: {hum_relative_error:.2f}%")

# Print first 5 point-wise errors
print("\nFirst 5 Point-wise Relative Errors:")
print(f"Temperature: {temp_point_errors[:5].values}")
print(f"Humidity: {hum_point_errors[:5].values}")
```
